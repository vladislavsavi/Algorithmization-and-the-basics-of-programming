<b>Конверация в числовые типы данных</b><br>
<i>Синтаксис</i><br>
<code>int(строка из цифр)</code> - введенная строка превратится в целочисленный тип данных, введенная строка не должна иметь никакие другие символы кроме цифр. int() также может действительное число превратить в целое число<br>

    <li><code>a = input("Enter some number"))

    <li><code>a = float(input("Enter a number without dot"))</code><br></li>

<h3>Загрузка модулей(библиотек)</h3>
<i>Синтаксис</i><br>
1. <code>import имя-модуля</code> - загружает модуль в текуший файл кода, методы, функций и константы доступны через имя-модуля<br>
2. <code>import имя-модуля as новое-кличка-модуля</code>- загружает модуль в текуший файл кода, методы, функций и константы доступны через новое-кличка-модуля<br>

<b>Пример:</b><br>
<code>import math
print(math.sin(10))</code><br>
Ouput: -0.5440211108893698

<h3>Основные функций модуля math</h3>
<ol>
    <li><code>math.ceil(x)</code> – округление вверх заданного числа x, не меняет значение</li>
    <li><code>math.floor(x)</code> – округление вниз заданного числа x, не меняет значение</li>
    <li><code>math.copysign(x,y)</code> – копирование знака y к x заданного числа x и y, не меняет значение</li>
    <li><code>math.fabs(x)</code> – абсолютное значение заданного числа x, не меняет значение</li>
    <li><code>math.factorial(x)</code> – факториал заданного числа x, не меняет значение</li>
    <li><code>math.log(x,y)</code> – y логарифм x, если не указан у, то это 2-ой логарифм, не меняет значение</li>
    <li><code>math.sin(x)</code> - x-радиан, sin заданного числа x, не меняет значение</li>
    <li><code>math.cos(x)</code> - x-радиан, cos заданного числа x, не меняет значение</li>
    <li><code>math.asin(x)</code> - x-радиан, arcsin заданного числа x, не меняет значение</li>
    <li><code>math.acos(x)</code> - x-радиан, arccos заданного числа x, не меняет значение</li>
    <li><code>math.tan(x)</code> - x-радиан, tan заданного числа x, не меняет значение</li>
    <li><code>math.atan(x)</code> - x-радиан, arctan заданного числа x, не меняет значение</li>
    <li><code>math.degrees(x)</code> – вывод градус угла из x-радиан</li>
    <li><code>math.radians(x)</code> – вывод радиан угла из x-градус</li>
    <li><code>math.pi</code> – число пи</li>
    <li><code>math.e</code> – число е</li>
</ol>
