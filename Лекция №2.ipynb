<h3>Логические операторы</h3>
<code>and</code> – И – Проверяет оба Булеан типы true<br>
<code>or</code> – ИЛИ – Проверяет хотябы один из Булеан типов true<br>
<code>not</code> – НЕТ – Делает обратным значение Булеан типа<br>

<b>Примеры</b><br>
<code>true and true</code> Output: true # true and true and true<br>
<code>true and false</code> Output: false # false or (true and false)<br>
<code>false and false</code> Output: false<br>
<code>true or true</code> Output: true<br>
<code>true or false</code> Output: true<br>
<code>false or false</code> Output: false<br>
<code>not true</code> Output: false<br>
<code>not false</code> Output: true<br>

<b>Вложенные условия</b><br>
Как говорилось ранее, условные операторы могут иметь все что было пройдена до, и могут иметь то что будем после проходить. Это озночает мы можем иметь вложенные условия.<br>

<b>Пример:</b>

```python
a = int(input())
if a>0:
    if a%2==0:
        print("a is even positive integer number")
    else:
        print("a is odd positive integer number")
else:
    if a%2==0:
        print("a is even negative integer number")
    else:
        print("a is odd negative integer number")
```
